```python
X=data.drop(labels=["price"],axis=1)
```

```python
Y=data[["price"]]
```

```python
cat_col=X.select_dtypes(include='object').columns
```

```python
num_col=X.select_dtypes(exclude='object').columns
```

```python
cut_map={'Premium':4, 'Very Good':3, 'Ideal':5, 'Good':2, 'Fair':1}
clarity_map={'VS2':4, 'SI2':2, 'VS1':5, 'SI1':3, 'IF':8, 'VVS2':6, 'VVS1':7, 'I1':1}
color_map={'F':3, 'J':7, 'G':4, 'E':2, 'D':1, 'H':5, 'I':6}
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
clarity_cat=['VS2', 'SI2', 'VS1', 'SI1', 'IF', 'VVS2', 'VVS1', 'I1']
cut_cat=['Premium', 'Very Good', 'Ideal', 'Good', 'Fair']
color_cat=['G', 'F', 'H', 'E', 'I', 'J', 'D'] 
```

```python
num_pipe=Pipeline(
steps=[
('imputer',SimpleImputer()),
('scaler',StandardScaler())

]
)
```

```python
cat_pipe=Pipeline(
    steps=[
('imputer',SimpleImputer(strategy='most_frequent')),
('encoding',OrdinalEncoder(categories=[cut_cat,color_cat,clarity_cat]))

]
)
```

```python
preprocesser=ColumnTransformer(
[
    ('num_pipeline',num_pipe,num_col),
    ('cat_pipeline',cat_pipe,cat_col)
]

)
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
preprocesser.fit_transform(X_train)
```

```python
preprocesser.transform(X_test)
```

```python
preprocesser.get_feature_names_out()
```

```python
X_train=pd.DataFrame(preprocesser.fit_transform(X_train),columns=preprocesser.get_feature_names_out())
```

```python
X_test=pd.DataFrame(preprocesser.transform(X_test),columns=preprocesser.get_feature_names_out())
```

```python
## Model Training

from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
## Train multiple models

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
```

```python
trained_model_list=[]
model_list=[]
r2_list=[]
```

```python
list(models)
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    print(model)
```

```python
models.keys()
```

```python
models.values()
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,Y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    #this is a validation(test) score
    mae, rmse, r2_square=evaluate_model(Y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
